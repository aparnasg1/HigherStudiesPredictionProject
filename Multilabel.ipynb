```python
import pandas as pd
import numpy as np

np.set_printoptions(precision=3, suppress=True)

import tensorflow as tf
from tensorflow.keras import layers
from tensorflow.keras.layers.experimental import preprocessing
from tensorflow.keras.layers import Dense, Input, Dropout,Flatten, Conv2D
```

```python
Admit_data = pd.read_csv(
    "Admission_Predict.csv",
    names=["GRE Score","TOEFL","University Rating","SOP","LOR","CGPA","Research", "Admit"])

Admit_data = Admit_data.drop("University Rating", 1)
# Admit_data = Admit_data.drop("Admit", 1)
Admit_data = Admit_data.iloc[1:]
Admit_data
```

```python
for i in ["GRE Score","TOEFL","SOP","LOR","CGPA","Research", "Admit"]:
    Admit_data[i] = Admit_data[i].astype(str).astype(float)
```

```python
X = Admit_data.copy()
Y = np.array(X.pop('Admit'))
```

```python
X = np.array(X)
```

```python
model = tf.keras.models.Sequential()

normalize = preprocessing.Normalization()
normalize.adapt(X)
model.add(normalize)

model.add(Dense(16, activation='relu',kernel_regularizer=tf.keras.regularizers.l1_l2(l1=1e-5, l2=1e-4),
    bias_regularizer=tf.keras.regularizers.l2(1e-4),
    activity_regularizer=tf.keras.regularizers.l2(1e-5)))
model.add(Dense(1, activation='sigmoid'))   # Final Layer using Softmax


```

```python
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
model.fit(X, Y, epochs=100)
model.summary()
```

```python
model.save('rec_model')
clg_list = pd.read_csv("CSV/clg_list.csv")
clg_list = np.array(clg_list)
```
