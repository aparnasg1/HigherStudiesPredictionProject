```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import BayesianRidge
import numpy as np
from sklearn import linear_model
import tkinter as tk 
import matplotlib.pyplot as plt
from matplotlib.backends.backend_tkagg import FigureCanvasTkAgg


df=pd.read_csv("C:/Users/Aparna Nair/Downloads/graduate-admissions (2)/Admission_Predict_Ver1.1.csv")

X=df[['GRE Score','TOEFL Score','University Rating', 'SOP', 'LOR ', 'CGPA',
       'Research']].astype(float)
Y=df['Chance of Admit '].astype(float) # output variable (what we are trying to predict)
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,random_state=11)

# with sklearn
lregr = linear_model.LinearRegression()
lregr.fit(X_train, Y_train)

# print('Intercept: \n', regr.intercept_)
# print('Coefficients: \n', regr.coef_)

bregr = linear_model.BayesianRidge()
bregr.fit(X_train, Y_train)

aregr = linear_model.ARDRegression()
aregr.fit(X_train, Y_train)

tregr = linear_model.TheilSenRegressor()
tregr.fit(X_train, Y_train)
# tkinter GUI
root= tk.Tk()

canvas1 = tk.Canvas(root, width = 500, height = 400)
canvas1.pack()

# with sklearn
# Intercept_result = ('Intercept: ', regr.intercept_)
# label_Intercept = tk.Label(root, text=Intercept_result, justify = 'center')
# canvas1.create_window(260, 280, window=label_Intercept)

# with sklearn
# Coefficients_result  = ('Coefficients: ', regr.coef_)
# label_Coefficients = tk.Label(root, text=Coefficients_result, justify = 'center')
# canvas1.create_window(260, 310, window=label_Coefficients)

def only_numbers(char):
    if(char.isdigit()):
        return True
    elif(char=='.'):
        return True
    else:
        return False
    
    
validation = root.register(only_numbers)

label1 = tk.Label(root, text='GRE Score: ')
canvas1.create_window(100, 100, window=label1)

entry1 = tk.Entry (root, validate="key", validatecommand=(validation, '%S')) # create 1st entry box
canvas1.create_window(270, 100, window=entry1)


label2 = tk.Label(root, text='TOEFL Score: ')
canvas1.create_window(100, 120, window=label2)

entry2 = tk.Entry (root, validate="key", validatecommand=(validation, '%S')) # create 2nd entry box
canvas1.create_window(270, 120, window=entry2)

label3 = tk.Label(root, text='University Rating: ')
canvas1.create_window(100, 140, window=label3)

entry3 = tk.Entry (root, validate="key", validatecommand=(validation, '%S')) # create 3rd entry box
canvas1.create_window(270, 140, window=entry3)


label4 = tk.Label(root, text='SOP: ')
canvas1.create_window(100, 160, window=label4)

entry4 = tk.Entry (root, validate="key", validatecommand=(validation, '%S')) # create 4th entry box
canvas1.create_window(270, 160, window=entry4)

label5 = tk.Label(root, text='LOR: ')
canvas1.create_window(100, 180, window=label5)

entry5 = tk.Entry (root, validate="key", validatecommand=(validation, '%S')) # create 5th entry box
canvas1.create_window(270, 180, window=entry5)

label6 = tk.Label(root, text='CGPA: ')
canvas1.create_window(120, 200, window=label6)

entry6 = tk.Entry (root, validate="key", validatecommand=(validation, '%S')) # create 6th entry box
canvas1.create_window(270, 200, window=entry6)

label7 = tk.Label(root, text='Any research papers(1:yes 0:no): ')
canvas1.create_window(120, 220, window=label7)

entry7 = tk.Entry (root, validate="key", validatecommand=(validation, '%S')) # create 7th entry box
canvas1.create_window(270, 220, window=entry7)


def values(): 
    global New_GRE #our 1st input variable
    New_GRE = float(entry1.get()) 
    
    global New_TOEFL #our 2nd input variable
    New_TOEFL = float(entry2.get()) 
    
    global New_UR #our 3rd input variable
    New_UR = float(entry3.get()) 
    
    global New_SOP #our 4th input variable
    New_SOP = float(entry4.get()) 
    global New_LOR #our 5th input variable
    New_LOR = float(entry5.get()) 
    
    global New_CGPA #our 6th input variable
    New_CGPA = float(entry6.get()) 
    global New_RS #our 7th input variable
    New_RS = float(entry7.get()) 
    
    
#for linear regression
def linear():
    Prediction_result  = ('Predicted Chance of admission: ',lregr.predict([[New_GRE ,New_TOEFL, New_UR , New_SOP, New_LOR, New_CGPA, New_RS]]))
    label_Prediction = tk.Label(root, text= Prediction_result, bg='orange')
    canvas1.create_window(260, 340, window=label_Prediction)
    #plot 1st scatter 
    figure3 = plt.Figure(figsize=(7,4), dpi=100)
    ax3 = figure3.add_subplot(111)
    ax3.scatter(df['GRE Score'].astype(float),df['Chance of Admit '].astype(float), color = 'r')
    scatter3 = FigureCanvasTkAgg(figure3, root) 
    scatter3.get_tk_widget().pack(side=tk.LEFT, fill=tk.BOTH)
    ax3.legend() 
    ax3.set_xlabel('GRE Score')
    ax3.set_title('GRE Score Vs. Chance of Admit ')

    #plot 2nd scatter 
    figure4 = plt.Figure(figsize=(7,4), dpi=100)
    ax4 = figure4.add_subplot(111)
    ax4.scatter(df['TOEFL Score'].astype(float),df['Chance of Admit '].astype(float), color = 'g')
    scatter4 = FigureCanvasTkAgg(figure4, root) 
    scatter4.get_tk_widget().pack(side=tk.RIGHT, fill=tk.BOTH)
    ax4.legend() 
    ax4.set_xlabel('TOEFL Score')
    ax4.set_title('TOEFL Score Vs. Chance of Admit ')


#for Bayesian ridge regression    
def bayesian():
    
    Prediction_result  = ('Predicted Chance of admission: ', bregr.predict([[New_GRE ,New_TOEFL, New_UR , New_SOP, New_LOR, New_CGPA, New_RS]]))
    label_Prediction = tk.Label(root, text= Prediction_result, bg='orange')
    canvas1.create_window(260, 340, window=label_Prediction)
    
#for ARD Regression
def ard():
    
    Prediction_result  = ('Predicted Chance of admission: ', aregr.predict([[New_GRE ,New_TOEFL, New_UR , New_SOP, New_LOR, New_CGPA, New_RS]]))
    label_Prediction = tk.Label(root, text= Prediction_result, bg='orange')
    canvas1.create_window(260, 340, window=label_Prediction)
    
#for TheilSen Regressor    
def theilsen():
    Prediction_result  = ('Predicted Chance of admission: ', tregr.predict([[New_GRE ,New_TOEFL, New_UR , New_SOP, New_LOR, New_CGPA, New_RS]]))
    label_Prediction = tk.Label(root, text= Prediction_result, bg='orange')
    canvas1.create_window(260, 340, window=label_Prediction)
    
button1 = tk.Button (root, text='Linear Regression',command=lambda:[values(), linear()], bg='orange') # button to call the 'values' command above 
canvas1.create_window(90, 380, window=button1)
button2 = tk.Button (root, text='Bayesian Ridge',command=lambda:[values(), bayesian()], bg='orange') # button to call the 'values' command above 
canvas1.create_window(200, 380, window=button2)
button3 = tk.Button (root, text='ARD Regression',command=lambda:[values(), ard()], bg='orange') # button to call the 'values' command above 
canvas1.create_window(300, 380, window=button3)
button4 = tk.Button (root, text='TheilSen Regression',command=lambda:[values(), theilsen()], bg='orange') # button to call the 'values' command above 
canvas1.create_window(410, 380, window=button4)

#  #plot 1st scatter 
# figure3 = plt.Figure(figsize=(3,2), dpi=100)
# ax3 = figure3.add_subplot(111)
# ax3.scatter(df['GRE Score'].astype(float),df['Chance of Admit '].astype(float), color = 'r')
# scatter3 = FigureCanvasTkAgg(figure3, root) 
# scatter3.get_tk_widget().pack(side=tk.LEFT, fill=tk.BOTH)
# ax3.legend() 
# ax3.set_xlabel('GRE Score')
# ax3.set_title('GRE Score Vs. Chance of Admit ')

# #plot 2nd scatter 
# figure4 = plt.Figure(figsize=(3,2), dpi=100)
# ax4 = figure4.add_subplot(111)
# ax4.scatter(df['TOEFL Score'].astype(float),df['Chance of Admit '].astype(float), color = 'g')
# scatter4 = FigureCanvasTkAgg(figure4, root) 
# scatter4.get_tk_widget().pack(side=tk.RIGHT, fill=tk.BOTH)
# ax4.legend() 
# ax4.set_xlabel('TOEFL Score')
# ax4.set_title('TOEFL Score Vs. Chance of Admit ')


root.mainloop()
```
