```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import BayesianRidge
import numpy as np
from sklearn import linear_model
import tkinter as tk 
from tkinter import messagebox
import tkinter.ttk as ttk
import tkvalidate as tkv
import matplotlib.pyplot as plt
from matplotlib.backends.backend_tkagg import FigureCanvasTkAgg

np.set_printoptions(precision=3, suppress=True)

import tensorflow as tf
from tensorflow.keras import layers
from tensorflow.keras.layers.experimental import preprocessing
from tensorflow.keras.layers import Dense, Input, Dropout,Flatten, Conv2D
model = tf.keras.models.load_model('rec_model')


df=pd.read_csv("Admission_Predict_Ver1.1.csv")

X=df[['GRE Score','TOEFL Score','University Rank', 'SOP', 'LOR ', 'CGPA',
       'Research']].astype(float)
Y=df['Chance of Admit '].astype(float) # output variable (what we are trying to predict)
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,random_state=11)

# with sklearn
lregr = linear_model.LinearRegression()
lregr.fit(X_train, Y_train)


root= tk.Tk()

root.title("WELCOME DEAR STUDENTS")

canvas1 = tk.Canvas(root, width = 700, height = 400)
canvas1.pack()



def only_numbers(char):
    if(char.isdigit()):
        return True
    elif(char=='.'):
        return True
    else:
        messagebox.showinfo('Information','Please enter only integer values')
        return False
    
    
validation = root.register(only_numbers)

label1 = tk.Label(root, text='GRE ',font=("Javanese Text", 10))
canvas1.create_window(100, 100, window=label1)

entry1 = tk.Entry (root, validate="key", validatecommand=(validation, '%S')) # create 1st entry box
canvas1.create_window(250, 100, window=entry1)

label2 = tk.Label(root, text='TOEFL',font=("Javanese Text", 10))
canvas1.create_window(400, 100, window=label2)

entry2 = tk.Entry (root, validate="key", validatecommand=(validation, '%S')) # create 2nd entry box
canvas1.create_window(550, 100, window=entry2)

label3 = tk.Label(root, text='UNIVERSITY RATING ',font=("Javanese Text", 10))
canvas1.create_window(100, 140, window=label3)

entry3 = tk.Entry (root, validate="key", validatecommand=(validation, '%S')) # create 3rd entry box
canvas1.create_window(250, 140, window=entry3)


label4 = tk.Label(root, text='SOP ',font=("Javanese Text", 10))
canvas1.create_window(400, 140, window=label4)

entry4 = tk.Entry (root, validate="key", validatecommand=(validation, '%S')) # create 4th entry box
canvas1.create_window(550, 140, window=entry4)

label5 = tk.Label(root, text='LOR ',font=("Javanese Text", 10))
canvas1.create_window(100, 180, window=label5)

entry5 = tk.Entry (root, validate="key", validatecommand=(validation, '%S')) # create 5th entry box
canvas1.create_window(250, 180, window=entry5)

label6 = tk.Label(root, text='CGPA ',font=("Javanese Text", 10))
canvas1.create_window(400, 180, window=label6)

entry6 = tk.Entry (root, validate="key", validatecommand=(validation, '%S')) # create 6th entry box
canvas1.create_window(550, 180, window=entry6)

label7 = tk.Label(root, text='RESEARCH PAPER PUBLISHED? ',font=("Javanese Text", 10))
canvas1.create_window(250, 220, window=label7)

label7 = tk.Label(root, text='ANY INTERNSHIP DONE? ',font=("Javanese Text", 10))
canvas1.create_window(250, 250, window=label7)

i = tk.IntVar() #Basically Links Any Radiobutton With The Variable=i.
r1 = tk.Radiobutton(root, text="YES", value=1, variable=i)
canvas1.create_window(380, 220, window=r1)
r2 = tk.Radiobutton(root, text="NO", value=0, variable=i)
canvas1.create_window(440, 220, window=r2)

j = tk.IntVar() #Basically Links Any Radiobutton With The Variable=i.
r1 = tk.Radiobutton(root, text="YES", value=1, variable=j)
canvas1.create_window(380, 250, window=r1)
r2 = tk.Radiobutton(root, text="NO", value=0, variable=j)
canvas1.create_window(440, 250, window=r2)
# entry7 = tk.Entry (root, validate="key", validatecommand=(validation, '%S')) # create 7th entry box
# canvas1.create_window(270, 220, window=entry7)


def values(): 
    global New_GRE #our 1st input variable
    New_GRE = float(entry1.get()) 
    
    global New_TOEFL #our 2nd input variable
    New_TOEFL = float(entry2.get()) 
    
    global New_UR #our 3rd input variable
    New_UR = float(entry3.get()) 
    
    global New_SOP #our 4th input variable
    New_SOP = float(entry4.get()) 
    global New_LOR #our 5th input variable
    New_LOR = float(entry5.get()) 
    
    global New_CGPA #our 6th input variable
    New_CGPA = float(entry6.get()) 
    global New_RS #our 7th input variable
    New_RS = float(i.get()) 
    
    
#for linear regression
def linear():
    Prediction_result  = ('Predicted Chance of admission: ',(lregr.predict([[New_GRE ,New_TOEFL, New_UR , New_SOP, New_LOR, New_CGPA, New_RS]]))*100)
    label_Prediction = tk.Label(root, text= Prediction_result, bg='orange')
    canvas1.create_window(300,300, window=label_Prediction)
    #plot 1st scatter 
    figure3 = plt.Figure(figsize=(7,4), dpi=100)
    ax3 = figure3.add_subplot(111)
    ax3.scatter(df['GRE Score'].astype(float),df['Chance of Admit '].astype(float), color = 'r')
    scatter3 = FigureCanvasTkAgg(figure3, root) 
    scatter3.get_tk_widget().pack(side=tk.LEFT, fill=tk.BOTH)
    ax3.legend() 
    ax3.set_xlabel('GRE Score')
    ax3.set_title('GRE Score Vs. Chance of Admit ')

    #plot 2nd scatter 
    figure4 = plt.Figure(figsize=(7,4), dpi=100)
    ax4 = figure4.add_subplot(111)
    ax4.scatter(df['TOEFL Score'].astype(float),df['Chance of Admit '].astype(float), color = 'g')
    scatter4 = FigureCanvasTkAgg(figure4, root) 
    scatter4.get_tk_widget().pack(side=tk.LEFT, fill=tk.BOTH)
    ax4.legend() 
    ax4.set_xlabel('TOEFL Score')
    ax4.set_title('TOEFL Score Vs. Chance of Admit ')

def topten():
    nayaWindow=tk.Toplevel(root)
    nayaWindow.title("TOP 10 UNIVERSITIES")
    nayaWindow.geometry("650x330")
    
    style= ttk.Style()

    style.configure("Treeview",rowheight=30)
    style.map('Treeview',
             background=[('selected','orange')])
    my_tree = ttk.Treeview(nayaWindow)
    my_tree['columns'] = ("GRE SCORE","UNIVERSITY RATING","UNIVERSITY NAME")

    # Formate Our Columns
    my_tree.column("#0", width=0, stretch='no')
    my_tree.column("#1", width=100, anchor='w')
    my_tree.column("#2", width=140,anchor='center')
    my_tree.column("#3", width=400, anchor='w')

    # Create Headings 
    my_tree.heading("#0", text="", anchor='w')
    my_tree.heading("#1", text="GRE SCORE", anchor='w')
    my_tree.heading("#2", text="UNIVERSITY RATING", anchor='center')
    my_tree.heading("#3", text="UNIVERSITY NAME",anchor='w')
    
    data = [ ["340", "5", "Stanford University"],
         ["340", "5", "Massachusetts Institute Of Technology"],
         ["335", "5", "Princeton University"],
         ["334", "4", "Carnegie Mellon University"],
         ["331", "4", "University Of California Berkeley"],
         ["339", "3", "University At Illinois At Urbana-Champaign"] ,
         ["336", "4", "Cornell University"],
         ["333", "3", "University At North-Carolina"],
         ["339", "4", "Harvard University"],
         ["336", "3", "University Of California-LA"],
         ["333", "3", "University Of California-Santa Barbara"],
         ["328", "3", "Georgia Institute Of Technology"] ]
    
    my_tree.tag_configure('oddrow',background='#ffffff')
    my_tree.tag_configure('evenrow',background='#f2f2f2')
    global count
    count=0

    for record in data:
        if count % 2 == 0:
            my_tree.insert(parent='', index='end', iid=count, text="", values=(record[0], record[1], record[2]), tags=('evenrow',))
        else:
            my_tree.insert(parent='', index='end', iid=count, text="", values=(record[0], record[1], record[2]), tags=('oddrow',))

        count += 1
        
    my_tree.pack()

def recommend():
    clg_list = pd.read_csv("CSV/clg_list.csv")
    clg_list = np.array(clg_list)
    X_test = np.array([New_GRE ,New_TOEFL, New_SOP, New_LOR, New_CGPA, New_RS])
    p = float(model.predict(np.array([X_test])))
    
    newWindow=tk.Toplevel(root)
    newWindow.title("PREDICTED UNIVERSITIES")
    newWindow.geometry("470x330")
    
    style= ttk.Style()

    style.configure("Treeview",rowheight=30)
    style.map('Treeview',
             background=[('selected','orange')])

    tree = ttk.Treeview(newWindow, columns=('SR NO.','UNIVERSITY'))
    tree.heading('#0', text='SR NO.')
    tree.heading('#1', text='UNIVERSITY')
    tree.column('#0', width=60, anchor='w')
    tree.column('#1', width=400)
    tree.tag_configure('oddrow',background='#ffffff')
    tree.tag_configure('evenrow',background='#f2f2f2')
    
    j = 1
    for i in clg_list:
        if p*100 >= float(i[2]):
#             lb.insert("end",i[1])
            if j%2==0:
                tree.insert('', 'end', text=j,values=(i[1],), tags=('evenrow',))
            else:
                tree.insert('', 'end', text=j,values=(i[1],), tags=('oddrow',))
            j += 1
            if j >= 11:
                break   
#     lb.pack()
    tree.pack()
    

button1 = tk.Button (root, text='CHANCES OF ADMISSION',command=lambda:[values(), linear()], bg='orange') # button to call the 'values' command above 
canvas1.create_window(90, 340, window=button1)
 


button2 = tk.Button(root,text="PREDICT UNIVERSITIES",command=lambda:[values(), recommend()],bg='orange')
canvas1.create_window(300, 380, window=button2)


button3 = tk.Button(root, text="TOP 10 UNIVERSITIES",command = topten,bg='orange')
canvas1.create_window(480, 340, window=button3)

root.mainloop()  
```
